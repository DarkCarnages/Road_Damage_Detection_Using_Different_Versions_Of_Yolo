```python

from ultralytics import YOLO


data_yaml = "/content/dataset/data.yaml"

model = YOLO("/content/yolov8s.pt")


model.train(
    data=data_yaml,
    epochs=50,
    imgsz=640,
    batch=16,

    name="road_damage_yolov8"
)


metrics = model.val()
print(metrics)


```

```python
import os
from ultralytics import YOLO
from google.colab import files

best_model_path = "/content/runs/detect/road_damage_yolov8/weights/best.pt"
last_model_path = "/content/runs/detect/road_damage_yolov8/weights/last.pt"
data_yaml = "/content/dataset/data.yaml"

results_root = "/content/results"
os.makedirs(results_root, exist_ok=True)


def evaluate(model_path, name):
    print(f"\n=== Evaluating {name}.pt ===")

    save_dir = f"{results_root}/{name}"
    os.makedirs(save_dir, exist_ok=True)

    model = YOLO(model_path)


    results = model.val(
        data=data_yaml,
        save=True,
        save_json=True,
        save_hybrid=True,
        plots=True,
        project=save_dir,
        name="eval"
    )


    metrics = {
        "Precision": float(results.box.mp),
        "Recall": float(results.box.mr),
        "mAP50": float(results.box.map50),
        "mAP50-95": float(results.box.map),
        "Fitness": float(results.fitness)
    }


    with open(f"{save_dir}/eval/metrics.txt", "w") as f:
        for k, v in metrics.items():
            f.write(f"{k}: {v}\n")

    print(f"âœ” Results saved at {save_dir}/eval/\n")


evaluate(best_model_path, "best")
evaluate(last_model_path, "last")

zip_path = "/content/results.zip"
!zip -r {zip_path} {results_root}

print("\nğŸ“¦ Results zipped successfully.")

files.download(zip_path)
```
